```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
url = 'https://drive.google.com/uc?id=1id___NYaSi1FYR8N_12N-hCFtB2pJJs7'
```

```python
population = pd.read_csv(url)
population
```

```python
population.info()
```

```python
#make var top_10 sort based on "2022 Population"
top_10 = population.sort_values(by='2022 Population', ascending=False).head(10)
top_10
```

```python
#line chart top_10['Country/Territory'],top_10['2022 Population']
plt.figure(figsize=(12,6))
plt.plot(top_10['Country/Territory'],top_10['2022 Population'],
         marker='d',
         color="red")
plt.title('Top 10 Population')
plt.xlabel('Country/Territory')
plt.ylabel('2022 Population')
plt.grid(True)
plt.show()
```

```python
population_saries = top_10.iloc[1,5:12]
```

```python
population_saries
```

```python
population_saries.index
```

```python
#make bar chart population_saries
plt.figure(figsize=(12,6))
plt.bar(population_saries.index,population_saries.values,color='red')
```

```python
population_series = top_10.iloc[:2,5:12]
```

```python
population_series
```

```python
diff = population_series.iloc[0,::] - population_series.iloc[1,::]
```

```python
diff
```

```python
plt.figure(figsize=(12,6))
plt.plot(diff.index,diff,color='blue',marker='d',markersize=8,
         markerfacecolor='lavender',markeredgecolor='black')
plt.bar(diff.index,diff,width=0.4,color='red',label='difference')
plt.title('Difference between the population of China and India ',size=24,color='lime')
plt.xlabel('years',size=24,color='#3A614B')
plt.ylabel('difference b/w population',size=24,color='#3A614B')
plt.show()
```
