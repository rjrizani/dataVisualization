```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://drive.google.com/uc?id=1Ld-ez8xqx-ntpE7JgBZpSsECrrAOkRSR'
```

```python
titanic_data = pd.read_csv(url)
titanic_data.head()
```

```python
titanic_data.info()
```

```python
titanic_data.drop("Cabin", axis=1, inplace=True)
```

```python
titanic_data.dropna(inplace=True)
titanic_data.reset_index(drop=True, inplace=True)
```

```python
titanic_data.info()
```

```python
titanic_data
```

```python
titanic_data.describe()
```

```python
survived_count = titanic_data['Survived'].value_counts()
survived_count
```

```python
plt.figure(figsize=(8, 6))
plt.pie(survived_count, labels=['Not Survived', 'Survived'], autopct='%1.1f%%', startangle=140)
plt.title('Survived vs Not Survived')
plt.show()
```

```python
sns.boxplot(data=titanic_data, x='Pclass', y='Age')
plt.show()
```

```python
#buat countplot  x= Pclass
sns.countplot(data=titanic_data, x='Pclass', hue = 'Survived', color='pink')
plt.show()
```
