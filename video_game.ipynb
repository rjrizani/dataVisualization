```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
url  = 'https://drive.google.com/uc?id=1NI1zHuMTh6C9LY20KcM2DHBCXyHODbxH'
```

```python
vg_game = pd.read_csv(url)
vg_game.head()
```

```python
vg_game.tail()
```

```python
vg_game['Global_Sales'].sort_values(ascending=False)[0:10]
```

```python
vg_game[vg_game['Global_Sales'] > 28.31]
```

```python
top_10 = vg_game[vg_game['Global_Sales'] > 28.31]
plt.bar(top_10['Genre'], top_10['Global_Sales'])
plt.xlabel('Genre')
plt.ylabel('Global Sales')
plt.title('Top 10 Games by Global Sales')
plt.show()
```

```python
vg_game.dropna(inplace=True)
```

```python
vg_game.isnull().sum()
```

```python
vg_game
```

```python
vg_game.reset_index(drop=True, inplace=True)
```

```python
vg_game[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].dtypes
```

```python
vg_game['NA_Sales'] = pd.to_numeric(vg_game['NA_Sales'])
vg_game['EU_Sales'] = pd.to_numeric(vg_game['EU_Sales'])
vg_game['JP_Sales'] = pd.to_numeric(vg_game['JP_Sales'])
vg_game['Other_Sales'] = pd.to_numeric(vg_game['Other_Sales'])
vg_game['Global_Sales'] = pd.to_numeric(vg_game['Global_Sales'])
```

```python
vg_game.fillna(0, inplace=True)
```

```python
print(vg_game[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']].head())
```

```python
vg_game['Total_Sales'] = vg_game['NA_Sales'] + vg_game['EU_Sales'] + vg_game['JP_Sales'] + vg_game['Other_Sales'] + vg_game['Global_Sales']
```

```python
print(vg_game['Total_Sales'].head())
```

```python
vg_game
```

```python
vg_game['Total_Sales'].sort_values(ascending=False)[:10]
```

```python
#di atas 56.62
top_10_games = vg_game[vg_game['Total_Sales']>=56.62]
top_10_games
```

```python
#x=Name, y=Total_Sales
```

```python
plt.figure(figsize=(16,7))
plt.xticks(rotation=20)
sns.barplot(x='Name', y='Total_Sales', data=top_10_games)
plt.title('Top 10 Games by Total Sales')
plt.show()
```
