# **Pie Chart**

```python
labels = ['Python','C++','Ruby','Java']
sizes = [215,130,245,210]
colors = ['gold','yellowgreen','lightcoral','lightskyblue']
plt.pie(sizes, labels=labels, autopct='%.1f%%',
        explode=[0,0.5,0.3,0], shadow=True, startangle=45,
        colors= colors)

plt.axis('equal')
plt.legend()
plt.title('Pie Chart')
plt.show()
```

#**Compare Plots**

```python
x = np.linspace(0,10,360)
```

# **SubPlots**

```python
plt.figure(figsize=(10,5))
plt.suptitle("2 subplot")

sub = ['math','science','lang','bio']
m1 =[80,70,23,56]
m2 = [12,34,54,56]

plt.subplot(2,1,1)
plt.bar(sub,m1, label='bar1')
plt.title('bar1')
plt.legend()

plt.subplot(2,1,2)
plt.bar(sub,m2, label='bar2')
plt.title('bar2')
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
activities = ['eat','sleep','work','play','other']
time_spent = [3,12,6,3,5.5]
colors = ['c', 'g', 'y','m', 'hotpink', '#32a8a8']

#create a figure with subplots
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10,8))

#plot on etch subplot
axs[0,0].plot(activities, time_spent)
axs[0,0].set_title('subplot1 line graph')

axs[0,1].bar(activities, time_spent)
axs[0,1].set_title('subplot2 bar graph')

axs[1,0].scatter(activities, time_spent)
axs[1,0].set_title('subplot3 scatter graph')

axs[1,1].pie(time_spent, labels=activities, autopct='%.1f%%', colors = colors)
axs[1,1].set_title('subplot4 pie chart')

plt.tight_layout()
plt.show()
```
