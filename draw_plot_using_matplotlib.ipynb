```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = list(range(1,6))
print(x)
```

```python
y = list(range(1,10,2))
print(y)
```

```python
plt.figure(figsize=(10,5))
plt.bar(x,y)

#add label
plt.xlabel('x')
plt.ylabel('y')
plt.title('Bar Chart')
plt.show()
```

```python
activities = ['eat','sleep','work','play','other']
time_spent = [3,12,6,3,5.5]

plt.figure(figsize=(10,5))
plt.bar(activities,time_spent)
plt.xlabel('Activity')
plt.ylabel('Time spent')
plt.title('Activity Chart')
plt.show()
```

```python
activities = ['eat','sleep','work','play','other']
time_spent = [3,12,6,3,5.5]

plt.figure(figsize=(10,5))
plt.barh(activities,time_spent)
plt.xlabel('Activity')
plt.ylabel('Time spent')
plt.title('Activity Chart')
plt.show()
```

```python
activities = ['eat','sleep','work','play','other']
time_spent = [3,12,6,3,5.5]

plt.figure(figsize=(8,4))
plt.bar(activities,time_spent, color='#32a8a8', width=0.2)
plt.xlabel('Activity')
plt.ylabel('Time spent')
plt.title('Activity Chart')
plt.show()
```

```python
bulan = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
pendapatan = [15000,18000,20500,22000,19500,25000,28000,24500,23000,26000,30000,32500]

plt.figure(figsize=(10,5))
plt.bar(bulan,pendapatan,width = 0.5)

#add label
plt.xlabel('bulan')
plt.ylabel('pendapatan(USD)')
plt.title('pendapatan Chart')

#for setting x labels vertical
plt.xticks(range(len(bulan)),bulan,rotation=45)

plt.show()
```

# **HISTOGRAM**

```python

marks = np.random.randint(0,100,150)
print(marks)
```

```python
print(len(marks))
```

```python
plt.xlabel('Marks')
plt.ylabel('Number of student')
plt.title('Number of students vs marks')
plt.hist(marks, bins=10, edgecolor='green')
plt.grid(True)
```

# **Pie Chart**

```python
labels = ['Python','C++','Ruby','Java']
sizes = [215,130,245,210]
colors = ['gold','yellowgreen','lightcoral','lightskyblue']
plt.pie(sizes, labels=labels, autopct='%.1f%%',
        explode=[0,0.5,0.3,0], shadow=True, startangle=45,
        colors= colors)

plt.axis('equal')
plt.legend()
plt.title('Pie Chart')
plt.show()
```
